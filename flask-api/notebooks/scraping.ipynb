```python
import requests
from bs4 import BeautifulSoup
import csv


url = 'https://www.amazon.com'


response = requests.get(url)


if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    products = soup.find_all('div', class_='product-item')  
    
 
    with open('/data/amazon_products_1mY.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        

        writer.writerow([
            'asin', 'title', 'imgUrl', 'productURL', 'stars', 'reviews', 
            'price', 'listPrice', 'category_id', 'isBestSeller', 'boughtInLastMonth'
        ])

        for product in products:

            asin = product.get('data-asin', 'N/A')
            
        
            title = product.find('h2', class_='product-title').text.strip()  
         
            img_tag = product.find('img')
            imgUrl = img_tag['src'] if img_tag else 'N/A'
      
            productURL = product.find('a')['href'] if product.find('a') else 'N/A'
            
            
            stars_tag = product.find('span', class_='stars') 
            stars = stars_tag.text.strip() if stars_tag else 'N/A'
            
            reviews_tag = product.find('span', class_='reviews') 
            reviews = reviews_tag.text.strip() if reviews_tag else 'N/A'
            
      
            price_tag = product.find('span', class_='price')  
            price = price_tag.text.strip() if price_tag else 'N/A'

            listPrice_tag = product.find('span', class_='list-price')  
            listPrice = listPrice_tag.text.strip() if listPrice_tag else 'N/A'
            
   
            category_id = product.get('data-category-id', 'N/A')
            
       
            isBestSeller = 'Best Seller' in product.text  
            
        
            boughtInLastMonth_tag = product.find('span', class_='bought-last-month')  
            boughtInLastMonth = boughtInLastMonth_tag.text.strip() if boughtInLastMonth_tag else 'N/A'
            

            writer.writerow([
                asin, title, imgUrl, productURL, stars, reviews, 
                price, listPrice, category_id, isBestSeller, boughtInLastMonth
            ])
    
    print("Data has been successfully scraped and saved to 'products.csv'.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
